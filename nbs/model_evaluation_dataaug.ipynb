```python
import pickle
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from train_dataaug import test
from datasets_dataaug import IEMOCAPEval
#os.environ['CUDA_VISIBLE_DEVICES'] = '2'
```

## to see the results at once

```python
from glob import glob
test_dataset = IEMOCAPEval(partition='test')
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False, num_workers=1, pin_memory=True)

for date, ratio in zip(['20220506_202846', '20220506_202925', '20220506_154958', '20220506_202913', '20220430_055023'], ['0.1', '0.25', '0.5', '0.75', '1.0']):
    for trial in range(5):
        print("############", ratio, trial, '##################')
        CHECKPOINT_MODEL_DIR = sorted(glob("checkpoint/*/model_fairness_dataaug_ratio{}_trial{}*.pth".format(ratio, trial)))
        print(CHECKPOINT_MODEL_DIR)
        model = torch.load(CHECKPOINT_MODEL_DIR[-1]).cuda()
        criterion = nn.CrossEntropyLoss()
        test(model, test_loader, test_dataset, criterion, return_fairness_eval=True)
```
